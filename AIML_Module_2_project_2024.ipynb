```python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data = onehot_encode(
    data,
    columns=['Beverage', 'Beverage_prep'],
    prefixes=['bev', 'bevp']
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler


data = pd.read_csv("heart.csv")
X = data.drop(columns=["target"])
y = data["target"]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)


tsne_df = pd.DataFrame(X_tsne, columns=["TSNE 1", "TSNE 2"])
tsne_df["Target"] = y
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(tsne_df["TSNE 1"], tsne_df["TSNE 2"], c=tsne_df["Target"], cmap="coolwarm", alpha=0.7)
plt.xlabel("TSNE Component 1")
plt.ylabel("TSNE Component 2")
plt.title("t-SNE Visualization of Heart Dataset")
plt.colorbar(scatter, label="Target Label")
plt.grid()
plt.show()
```

Observations from the t-SNE Plot
Clusters Formation: If the plot shows well-separated clusters, it indicates that the dataset has clear underlying structures.
Overlap: If the clusters overlap, the data may not have strong separability using this method.
Non-Linear Structures: t-SNE preserves local relationships, so points close in high-dimensional space remain close in 2D.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv("heart.csv")


corr_matrix = data.corr()


print(corr_matrix)
```

```python

plt.figure(figsize=(10, 8))


sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)


plt.title("Correlation Heatmap of Heart Dataset")
plt.show()
```

Observations from the Heatmap
Strong Positive Correlations: Look for values close to +1. Features with high correlation might be redundant.

Example: If cholesterol and age are highly correlated, one might be less important in predictive models.
Strong Negative Correlations: Look for values close to -1. This indicates an inverse relationship.

Example: Max heart rate (thalach) may be negatively correlated with age (older individuals tend to have lower max heart rates).
Weak Correlations: Values near 0 suggest little to no linear relationship.
